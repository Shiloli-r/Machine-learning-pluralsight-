The predict function is based off the line equation: y = x * w + b (b is y-intercept and w is slope)

```python
def predict(X, w, b):
    return X * w + b
```

```python
predict(14, 1.2, 12)
```

```python
# Broadcasting input values
X = np.array([14, 5, 7])
predict(X, 1.2, 12)
```

Function to calculate the loss (how wrong the line is at approximating the dataset). The mean squared error is the loss

```python
def loss(X, Y, w, b):
    predictions = predict(X, w, b)
    return np.average((predictions - Y) ** 2)
```

```python
# Reload the dataset
X, Y = np.loadtxt("pizza.txt", skiprows=1, unpack=True)
loss(X, Y, 1.2, 12)
```

```python
def train(X, Y, iterations, lr):
    w = b = 0
    for i in range(iterations):
        current_loss = loss(X, Y, w, b)
        print("Iteration %4d => Loss: %.6f" % (i, current_loss))

        if loss(X, Y, w - lr, b) < current_loss:
            w -= lr
        elif loss(X, Y, w + lr, b) < current_loss:
            w += lr
        elif loss(X, Y, w, b - lr) < current_loss:
            b -= lr
        elif loss(X, Y, w, b + lr) < current_loss:
            b += lr
        else:
            return w, b

    raise Exception("Couldn't find a result within %d iterations" % iterations)
```

```python
w, b = train(X, Y, 10000, 0.01)
```
